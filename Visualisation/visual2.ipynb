```python
# Area chart
# An area chart is really similar to a line chart, except that 
# the area between the x axis and the line is filled in with color or shading

temp = corona_virus.groupby('ObservationDate')['Confirmed'].sum().reset_index()
fig = plt.figure(figsize=(20,12))
plt.fill_between(temp['ObservationDate'][:30], temp['Confirmed'][:30], color='r')
plt.xticks(rotation=90)
plt.xlabel('Observation Date')
plt.ylabel('No of confirmed cases')

plt.show()
```

```python
# Stacked area chart

india = corona_virus[corona_virus['Country/Region']=='India'].groupby('ObservationDate')['Confirmed'].sum().reset_index()
us = corona_virus[corona_virus['Country/Region']=='US'].groupby('ObservationDate')['Confirmed'].sum().reset_index()
brazil = corona_virus[corona_virus['Country/Region']=='Brazil'].groupby('ObservationDate')['Confirmed'].sum().reset_index()

fig = plt.figure(figsize=(20,12))
plt.stackplot(temp['ObservationDate'][-50:],
              us['Confirmed'][-50:],
               brazil['Confirmed'][-50:],
               labels=['US', 'India', 'Brazil'])
plt.xticks(rotation=90)
plt.legend(loc='upper left')
plt.xlabel('Observation Date')
plt.ylabel('No of confirmed cases')

plt.show()
```

```python
# Lollipop plot
# A lollipop plot is an hybrid between a scatter plot and a
# barplot. It shows the relationship between a numerical 
# variable and another variable, numerical OR categorical.

temp = latest_covid_cases.sort_values('Confirmed', ascending=False)[['Country_Region', 'Confirmed']]
plt.stem(temp['Country_Region'][:10], temp['Confirmed'][:10])
plt.xticks(rotation=90)
plt.show()
```

```python
# Horizontal lollipop
plt.hlines(y=temp['Country_Region'][:10][::-1], xmin=0,
           xmax=temp['Confirmed'][:10][::-1], color='skyblue')

plt.plot(temp['Confirmed'][:10][::-1], temp['Country_Region'][:10][::-1], 'D')
plt.xticks(rotation=90)
plt.show()
```
