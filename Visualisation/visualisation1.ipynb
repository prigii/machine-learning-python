```python
# data visualisation using matplotlib

import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('countries.csv')
data.head()
plt.plot(data['year'], data['population'])

```

```python
kenya = data[data['country'] == 'Kenya']
plt.plot(kenya['year'], kenya['population'] / 10**6, label='Kenya', color='red')
```

```python
comoros = data[data['country'] == 'Comoros']
plt.plot(comoros['year'], comoros['population'] / 10**6, label='Comoros', color='red')
plt.title('Comoros Population')
plt.xlabel('Year')
plt.ylabel('Population (Millions)')
```

```python
plt.plot(kenya['year'], kenya['population'] / 10**6, label='Kenya')
plt.plot(comoros['year'], comoros['population'] / 10**6, label='Comoros')
plt.legend(['Kenya', 'Comoros'])
plt.show()
```
