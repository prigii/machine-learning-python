```python
 # Scatter chart using pairs of points
 import matplotlib.pyplot as plt
 from pylab import randn
 
 x, y = randn(2, 100)
 plt.scatter(x, y, color='red', marker='o')
 plt.xlabel('x')
 plt.ylabel('y')
 plt.title('Scatter plot pythonspot.com')
 plt.show()
```
